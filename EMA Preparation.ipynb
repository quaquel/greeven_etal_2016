## 2. Model stochasticity
The model contains several stochastic elements: the random distribution of the time horizon attribute among individuals, a stochastic factor for the occurrence of a climate disaster and a stochastic factor for the locality of a climate disaster. 

The first part of this section is used to demonstrate that no stochasticity will occur if the stochastic elements are left out of the model, that is, if the climate disaster structure is left out of the model and the standard deviation of the time horizon distribution among individuals is set to 0. The graph illustrates the cumulative GHG reduction outcome of 100 experiments, which implies that all 100 experiments have an equal outcome.

The second part of this section is used to illustrate the influence of stochasticity in the model. 1000 experiments have been conducted at standard parameter settings of the model. The cumulative GHG reduction is used as the outcome of interest. The spread in outcomes is an illustration of the model stochasticity.

```python
results = load_results(r'./data/EMA results ModelSebastiaanGreeven 1000 exp Stoch Test.tar.gz')

experiments, outcomes = results

data = outcomes['CumulativeGHGreduction']

fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(np.mean(data, axis=0), label = 'mean')
ax.fill_between(np.arange(0, data.shape[1]),np.min(data, axis=0), np.max(data, axis=0), alpha=0.2)
ax.plot(np.percentile(data, 25, axis=0), label='first quartile')
ax.plot(np.percentile(data, 75, axis=0), label='third quartile')

ax.legend()
plt.show()
```

## 3. Determining required number of replications per model run
To correct for stochastic uncertainty in the model, each model run needs to be replicated. The stochasticity of the model has been illustrated by the outcome range in the previous section by displaying the outcomes of 1000 experiments at the same standard model settings. 

In order to determine the amount of replications needed for each model run, the mean of the cumulative GHG reduction is plotted against the number of experiments. The number of replications can then be chosen according to a value of the number of experiments in which the plotted mean is stabilising. It can be observed that around 100 experiments, the mean starts to stabilise. 

This analysis has determined that each model run needs to be replicated 100 times in order to correct for stochasticity. In the differnt analysis that will follow in this research, a model run is represented by the mean of 100 replications of that model run. 

```python
nr_replications = data.shape[0]

means = []
for i in range(1,nr_replications):
    means.append(np.mean(data[0:i, -1]))
    
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(means)
ax.set_xlabel('nr of replications')
ax.set_ylabel('mean CumulativeGHGreduction')
plt.show()
```
